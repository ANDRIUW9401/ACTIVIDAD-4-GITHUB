```python
# Sistema de transporte simulado (tipo sistema Masivo #Transmilenio,#Mio,Etc)

grafo = {
 'Portal Norte': {'Calle 100': 5, 'Calle 85': 7},
    'Calle 100': {'Portal Norte': 5, 'Calle 72': 4},
    'Escuela Militar': {'Calle 85': 2,'Calle 100': 5,},
    'Calle 85': {'Portal Norte': 7, 'Calle 72': 3},
    'Calle 72': {'Calle 100': 4, 'Calle 85': 3, 'Calle 26': 6},
    'Calle 26': {'Calle 72': 6, 'Portal Sur': 8},
    'Portal Sur': {'Calle 26': 8}
}
```

```python
# Estimación de distancia al destino
heuristica = {
    'Portal Norte': 15,
    'Calle 100': 10,
    'Escuela Militar': 8,
    'Calle 85': 9,
    'Calle 72': 6,
    'Calle 26': 3,
    'Portal Sur': 0
}
```

```python
def a_estrella(inicio, objetivo):
    abiertos = [inicio]
    cerrados = []
    costo = {inicio: 0}
    padre = {inicio: None}

    while abiertos:
        nodo = min(abiertos, key=lambda x: costo[x] + heuristica[x])

        if nodo == objetivo:
            camino = []
            while nodo:
                camino.append(nodo)
                nodo = padre[nodo]
            return camino[::-1]

        abiertos.remove(nodo)
        cerrados.append(nodo)

        for vecino, peso in grafo[nodo].items():
            nuevo_costo = costo[nodo] + peso

            if vecino not in abiertos and vecino not in cerrados:
                abiertos.append(vecino)
                padre[vecino] = nodo
                costo[vecino] = nuevo_costo

            elif nuevo_costo < costo.get(vecino, float('inf')):
                padre[vecino] = nodo
                costo[vecino] = nuevo_costo

    return None
```

```python
inicio =  'Portal Sur'
destino = 'Portal Norte'

ruta = a_estrella(inicio, destino)

print("Inicio:", inicio)
print("Destino:", destino)
print("Mejor ruta encontrada:", ruta)
```
